```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope,intercept,r,p,std_err=stats.linregress(x,y)

def myfunc(x):
    return slope*x+intercept

mymodel=list(map(myfunc,x))
print(mymodel)
print("relationship r value for the given arrays is: "+str(r))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()

speed=myfunc(10)
print(speed)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

df=pd.read_csv('data1.csv')
#print(df)
arr1=(df.iloc[:,0]).tolist()
print(arr1)
arr2=(df.iloc[:,1]).tolist()
print(arr2)

slope,intercept,r,p,std_err=stats.linregress(arr1,arr2)
print("r for the given data is: "+str(r)+"hence going for linear regression")
def myfunc(arr1):
    return slope*arr1+intercept

mymodel=list(map(myfunc,arr1))
plt.plot(arr1,mymodel)
plt.scatter(arr1,arr2)
plt.show()
speed=myfunc(1)
print("predicted speed for 1 is :"+str(speed))
```
